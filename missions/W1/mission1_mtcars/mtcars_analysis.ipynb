# 10. unique(), nunique(), value_counts()

### df[].unique()는 중복없이 출력
### df[].nunique()는 중복없는 값 개수
### value_counts()는 각 값이 몇개 있고 타입이 뭔지

```python
print(df["gear"].unique())
print(df["gear"].nunique())
print(df["gear"].value_counts())
print()
print(df["am"].unique())
print(df["am"].nunique())
print(df["am"].value_counts())
```

# 11. gear와 transmission

### 경우의 수를 조합별로 나누고 각 조합에 몇개의 차량이 있는지.

```python
print(df.groupby(["gear", "am"]).size())
print(df.groupby(["gear", "am"]).count())
```

# 12. 그래프 그리기

### 타이틀은 '# of Cars by Gear x Transmission'
### x축 레이블은 '(Gear, Auto Transmission)'
### y축 레이블은 '# of Cars'

```python
gear_am = df.groupby(["gear", "am"]).size()
print(gear_am)
```

### plot(kind='bar')
### 바그래프 그리기

```python
gear_am.plot(kind='bar')

plt.title("# of Cars by Gear x Transmission")
plt.xlabel("(Gear, Auto Transmission)")
plt.ylabel("# of Cars")

plt.show()
```

# 12. 데이터셋에 포함된 모든 features의 히스토그램

### hist() : 가능한 그래프 모두 그리기. model은 문자만 있어서 그래프 못그림
#### 그래프 타이틀은 "Histograms of Variables" : suptitle로 전체 그래프 제목

#### 최대한 잘 보이도록 그래프들 배치 : tight_layout + rect[위쪽에 5% 공간 남기기]

```python
df.hist(figsize=(12,10))
plt.suptitle("Histograms of Variables")
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

# 14. 상관계수 구하기

### 상관계수 : 두 변수의 사이 관계를 -1 ~ 1 로 표현
### 1 : 하나가 증가하면 다른 하나도 증가
### 0 : 거의 관계 없음
### -1 : 하나가 증가하면 다른 하나 감소

### df.corr()
### model은 표현 못해서 오류가 날 수 있어서 
### df.corr(numeric_only = True) 사용

### corr.round(n) : 소숫점 아래 n개 까지

```python
corr = df.corr(numeric_only = True)
corr
```

```python
corr2 = df.corr(numeric_only = True)
corr2.round(2)
```
