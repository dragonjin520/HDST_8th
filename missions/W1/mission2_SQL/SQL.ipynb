```python
conn = sqlite3.connect(":memory:")
cur = conn.cursor()
```

# 3. SQL 실행 함수 만들기

```python
def run_query(query):
    cur.execute(query)

    if cur.description:
        columns = [desc[0] for desc in cur.description]
        print(columns)

        rows = cur.fetchall()
        for row in rows:
            print(row)
    else:
        conn.commit()
        print("Query executed successfully.")
```

FROM table_name
WHERE condition
GROUP BY column1, column3
ORDER BY column_name;

FROM table_name
WHERE condition
GROUP BY column1, column3
HAVING condition -- The condition on grouped data
ORDER BY column_name;

### SQL ALL
W3Schools의 `= ALL (subquery)` 문법은 SQLite에서 직접 지원되지 않는다.  
또한 이 예제는 서브쿼리 결과의 모든 값이 동일해야만 참이 되므로, 실제 데이터에서는 결과가 없을 가능성이 높다.

# 36. SQL Stored Procedures
SQLserver

CREATE PROCEDURE procedure_name
  @param1 datatype,
  @param2 datatype
AS
BEGIN
  -- SQL_statements to be executed
  SELECT column1, column2
  FROM table_name
  WHERE columnN = @paramN;
END;

## SQL Stored Procedures / Drop Procedure

W3Schools의 `DROP PROCEDURE` 문법은 저장 프로시저를 삭제하는 명령어이다.  
하지만 SQLite는 Stored Procedure를 지원하지 않으므로,  
`CREATE PROCEDURE`, `EXEC`, `DROP PROCEDURE` 관련 예제는 실행하지 않고 문법과 개념만 확인하였다.

W3Schools의 Stored Procedure 예제는 특정 도시를 파라미터로 받아 해당 도시의 고객 정보를 조회하는 프로시저를 생성하고 실행하는 예제이다.  
하지만 SQLite는 `CREATE PROCEDURE`, `EXEC`, 파라미터 선언(`@City`)과 같은 Stored Procedure 문법을 지원하지 않는다.  
따라서 본 실습에서는 프로시저 자체를 생성하지 않고, 실제 동작에 해당하는 `SELECT * FROM Customers WHERE City = 'London';` 쿼리로 대체하여 실행하였다.
