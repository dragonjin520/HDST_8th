```python
conn = sqlite3.connect(":memory:")
cur = conn.cursor()
```

# 3. SQL 실행 함수 만들기

```python
def run_query(query):
    cur.execute(query)

    if cur.description:
        columns = [desc[0] for desc in cur.description]
        print(columns)

        rows = cur.fetchall()
        for row in rows:
            print(row)
    else:
        conn.commit()
        print("Query executed successfully.")
```
